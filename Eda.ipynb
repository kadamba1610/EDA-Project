```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = {
    "Date": ["2024-01-01","2024-01-02","2024-01-03","2024-01-04","2024-01-05",
             "2024-01-06","2024-01-07","2024-01-08","2024-01-09","2024-01-10"],
    "City": ["Chennai","Mumbai","Delhi","Chennai","Mumbai",
             "Delhi","Chennai","Mumbai","Delhi","Chennai"],
    "Product": ["Laptop","Phone","Tablet","Laptop","Phone",
                "Tablet","Laptop","Phone","Tablet","Laptop"],
    "Sales": [50000,30000,20000,55000,32000,21000,60000,35000,22000,58000],
    "Quantity": [5,10,7,6,11,8,7,12,9,6]
}

df= pd.DataFrame(data)
```

```python
df.groupby("City")["Sales"].sum()
```

```python
df.groupby("City")["Sales"].sum().plot(kind="bar",
color=["red","blue","green"])
plt.show()
```

```python
df.groupby("City")["Quantity"].sum().plot(kind="barh",color=["green","black","red"])
```

```python
df.groupby("City")["Sales"].max()
```

```python
df.groupby("Product")["Quantity"].sum()
```

```python
df.groupby("Product")["Quantity"].sum().plot(kind="pie",autopct="%1.1f%%")
plt.title("productdistribution")
plt.ylabel("")
```
